```python
#Importing required libraries
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB

#Importing the dataset
data = pd.read_csv("tennisdata.csv")
print("The 5 values of the data is:\n", data.head())

X = data.iloc[:, :-1]
print("The 5 values of the training data is:\n", X.head())

y = data.iloc[:, -1]
print("The 5 values of the training output is:\n", y.head())

#Converting them into numbers
le_outlook = LabelEncoder()
X.Outlook = le_outlook.fit_transform(X.Outlook)

le_temperature = LabelEncoder()
X.Temperature = le_temperature.fit_transform(X.Temperature)

le_humidity = LabelEncoder()
X.Humidity = le_humidity.fit_transform(X.Humidity)

le_windy = LabelEncoder()
X.Windy = le_windy.fit_transform(X.Windy)

print("Now the training output is:\n", X.head())

le_playtennis = LabelEncoder()
y = le_playtennis.fit_transform(y)
print("Now the training out is:\n",y)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.20)

classifer = GaussianNB()
classifer.fit(X_train, Y_train)

from sklearn.metrics import accuracy_score
print("Accuracy:\t", accuracy_score(classifer.predict(X_test), Y_test))
```
